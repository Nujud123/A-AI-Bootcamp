````python
def generate_password(length: int, lower: str, upper: str, digit: str, min_digit: int, symbol: str, min_symbol: int) -> str:
    password = ""
    i = 0
    
    for j in range(min_digit):
        if len(digit) > 0:
            password += random.choice(digit)
            i += 1
                
    for j in range(min_symbol):
        if len(symbol) > 0:
            password += random.choice(symbol)
            i += 1

    avaliable = lower + upper + digit + symbol

    while i < length:
        password += random.choice(avaliable)
        i += 1

    password = list(password)
    random.shuffle(password)
    password = "".join(password)
    return password
````

````python
print(generate_password(10, "abcdefghijklmnopqrstuvwxyz", "ABCDEFGHIJKLMNOPQRSTUVWXYZ", "0123456789", 2, "!@#$%^&*", 1))
````

````python
def generate_passphrase(words: list ,words_count: int, separator: str, capitalize: bool, add_digit: bool) -> str:
    passphrase = random.sample(words, words_count)

    if capitalize:
        for i in range(len(passphrase)):
            passphrase[i] = passphrase[i].capitalize()

    passphrase = separator.join(passphrase)

    if add_digit:
        passphrase += str(random.randrange(10))

    return passphrase
````

````python
words = ["apple", "river", "cloud", "sun", "book", "coffee", "moon", "tree"]
print(generate_passphrase(words, 3, "-", True, True))
````
