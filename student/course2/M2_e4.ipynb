### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')
chipo.head()
```

### Step 4. How many products cost more than $10.00?

```python
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))
```

```python
# solution 1
count = chipo['item_price'].apply(lambda x: 1 if x > 10 else 0)
count.sum()
```

```python
# solution 2
(chipo['item_price'] > 10).sum()
```

### Step 5. What is the price of each item? 
###### print a data frame with only three columns item_name choice_description and product_price

```python
# solution 1 "this print the three columns whithout any modifation"
item_prices = chipo[['item_name', 'choice_description', 'item_price']]
item_prices.head(10)
```

```python
# solution 2 "this print the mean price for each unique item"
item_prices = chipo.groupby(['item_name', 'choice_description'])[['item_price']].mean()
item_prices.head(10)
```

### Step 6. Sort by the name of the item

```python
# this dorted the items by name without take the mean price "just sort"
item_prices = chipo[['item_name', 'choice_description', 'item_price']]
item_prices = item_prices.sort_values(by='item_name')
item_prices.head(10)
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
# solution 1
expensive_item = chipo['item_price'].max()
chipo[chipo['item_price'] == expensive_item]['quantity'].iloc[0]
```

```python
# solution 2
chipo.loc[chipo['item_price'].idxmax(), 'quantity']
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
# solution 1
(chipo['item_name'] == 'Veggie Salad Bowl').sum()
```

```python
# solution 2
chipo[chipo['item_name'] == 'Veggie Salad Bowl'].shape[0]
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
ordered_soda = chipo[chipo['item_name'] == 'Canned Soda']
ordered_soda[ordered_soda['quantity'] > 1].shape[0]
```
